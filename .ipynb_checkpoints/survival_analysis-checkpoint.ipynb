## Survival Analysis: DirtSlurper3100

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt, seaborn as sns
from scipy.stats import chi2_contingency
from scipy.stats import mannwhitneyu
from lifelines import KaplanMeierFitter, CoxPHFitter, WeibullAFTFitter, LogNormalAFTFitter, LogLogisticAFTFitter

from lifelines.statistics import logrank_test
from datetime import datetime
import os

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv(os.path.join(os.getcwd(), "Dataset.csv"), sep=',')
print(f"Dataset contains {df.shape[0]} rows and {df.shape[1]} columns")
for column in df.columns:
     print(f"Name: {column:<25} Type: {type(column)}")
```

Registration dates in the dataset span from January 2015 to the end of December 2019. The Registration date column is currently stored as a string (ObjectDType), which we convert to a standard datetime format (dd/mm/yyyy) to enable duration calculations and date filtering. We apply the same transformation to the Failure date column.

Additionally, we convert the Pets and Sent for repair columns from strings to binary values (0 = no, 1 = yes) for easier computation. Placeholder values (---) are replaced with NaN to properly represent missing data.

```python
df.replace('---', np.nan, inplace=True)

df['Registration date'] = pd.to_datetime(df['Registration date'], format='%d%b%Y')
df["Failure date"] = df["Failure date"].replace("___", np.nan)
df["Failure date"] = pd.to_datetime(df["Failure date"], format='%d%b%Y', errors="coerce")

df["Pets"] = df["Pets"].map({"NO": 0, "YES": 1})
df["Sent for repair"] = df["Sent for repair"].map({"NO": 0, "YES": 1})
df["Battery status"] = df["Battery status"].map({"OK": 0, "Damage": 1})
df["Impact status"] = df["Impact status"].map({"OK": 0, "Damage": 1})
df["IR status"] = df["IR status"].map({"OK": 0, "Damage": 1})

df.head()
```

```python
censor_date = pd.to_datetime("31-12-2019", format="%d-%m-%Y")
df["Time"] = (df["Failure date"].fillna(censor_date) - df["Registration date"]).dt.days
df.head()
```

The total usage time of vacuum cleaners not sent for repair appears to follow a decreasing linear pattern, whereas vacuum cleaners sent for repair exhibit a more stable distribution with a more central mean and tighter interquartile range. This suggests that total usage time may be a significant factor. We will test for statistical dependence between total usage time and repair status using a Mann-Whitney U test.

Furthermore, households with pets are nearly twice as likely to have sent their vacuum for repair compared to households without pets, indicating that pet ownership may be a significant factor. We will test for statistical dependence between pet ownership and repair status using a chi-square test of independence.
Lastly, the carpet score distributions for both groups appear similar, suggesting no significant association. We will confirm this using a Mann-Whitney U test.

```python
factors = ['Total usage time', 'Pets', 'Carpet score']

df_all = pd.concat([df_sent_repair.assign(Repair=1), df_notsend_repair.assign(Repair=0)])

# p value matrix
p_matrix = pd.DataFrame(index=factors, columns=["Repair decision"], dtype=float)

for factor in factors:
    print("="*80)
    if factor == 'Total usage time' or factor == 'Carpet score':  # continuous use Mann Whitney U test
        group1 = df_notsend_repair[factor]
        group2 = df_sent_repair[factor]

        stat, p = mannwhitneyu(group1, group2, alternative='two-sided')
        p_matrix.loc[factor, "Repair decision"] = p

        print(f"Mann-Whitney U test for {factor} vs Repair decision")
        print("U statistic:", stat)
        print("p-value:", p)
        if p < 0.05:
            print(f"Reject H0: {factor} is significantly associated with sending for repair.")
        else:
            print(f"Fail to reject H0: No significant association.")

    else:  # chategorical use Chi square
        contingency = pd.crosstab(df[factor], df['Sent for repair'])
        chi2, p, dof, expected = chi2_contingency(contingency)
        p_matrix.loc[factor, "Repair decision"] = p

        print(f"Chi-square test for {factor} vs Repair decision")
        print("Chi-square statistic:", chi2)
        print("p-value:", p)
        print("Expected frequencies:\n", expected)
        if p < 0.05:
            print(f"Reject H0: {factor} is significantly associated with sending for repair.")
        else:
            print(f"Fail to reject H0: No significant association.")

# heatmap plot of p values
plt.figure(figsize=(5, 3))
sns.heatmap(p_matrix.astype(float), annot=True, fmt=".3f", cmap="coolwarm_r", cbar_kws={'label': 'p-value'}, linewidths=0.5, 
            linecolor='gray')

plt.title("P-value Matrix: Factors vs Repair Decision")
plt.tight_layout()
plt.show()
```

The distribution of registration dates shows that vacuums not sent for repair have linearly increasing registrations over time, while vacuums sent for repair exhibit linearly decreasing registrations. Given the statistically significant relationship between total usage time and repair status, these findings are expected: vacuums registered more recently have not had sufficient usage time to experience failure.

Failure dates increase linearly, confirming that failures occur throughout the product's lifecycle rather than being clustered at a specific point in time.

Having identified correlations between certain parameters, we now examine the characteristics of vacuum cleaners sent for repair more closely  specifically the IR status, battery status, and impact status. We therefore perform a conditional analysis on failures.
We test whether total usage time, pet ownership, and carpet score are significantly associated with battery status, impact status, and IR status  resulting in nine tests in total.

```python
factors = ['Total usage time', 'Pets', 'Carpet score']
components = ['Battery status', 'Impact status', 'IR status']

p_matrix = pd.DataFrame(index=factors, columns=components, dtype=float)

# loop over all pairs
for factor in factors:
    for component in components:
        print('='*80)
        if factor == 'Total usage time': # total usage time is continuous -> applying Mann Whitney U test
            print(f"# Null hypothesis H0: {factor} and {component} are independent (no difference in distributions).")
            
            group1 = df_sent_repair.loc[df_sent_repair[component] == 0, factor]
            group2 = df_sent_repair.loc[df_sent_repair[component] == 1, factor]
            
            stat, p = mannwhitneyu(group1, group2, alternative='two-sided')
            p_matrix.loc[factor, component] = p
            print(f"Mann-Whitney U test for {factor} vs {component}")
            print("U statistic:", stat)
            print("p-value:", p)
            if p < 0.05:
                print(f"Reject H0: {factor} is significantly associated with {component}.")
            else:
                print(f"Fail to reject H0: No significant association between {factor} and {component}.")
        
        else: # binary test applying chi 2 square test
            print(f"# Null hypothesis H0: {factor} and {component} are independent (no association).")
            
            contingency = pd.crosstab(df_sent_repair[factor], df_sent_repair[component])
            
            chi2, p, dof, expected = chi2_contingency(contingency)
            p_matrix.loc[factor, component] = p
            print(f"Chi-square test for {factor} vs {component}")
            print("Chi-square statistic:", chi2)
            print("p-value:", p)
            print("Expected frequencies:\n", expected)
            if p < 0.05:
                print(f"Reject H0: {factor} is significantly associated with {component}.")
            else:
                print(f"Fail to reject H0: No significant association between {factor} and {component}.")

plt.figure(figsize=(7, 4))
sns.heatmap(p_matrix.astype(float), annot=True, fmt=".3f", cmap="coolwarm_r", cbar_kws={'label': 'p-value'}, linewidths=0.5, 
            linecolor='gray')

plt.title("P-value Matrix: Factors vs Component Failures")
plt.xlabel("Component")
plt.ylabel("Factor")
plt.tight_layout()
plt.show()
```

These findings indicate that total usage time, pet ownership, and carpet score are significantly associated with most component statuses. The only exception is the relationship between carpet score and IR status, which shows no significant association.

## 2) Modelling

The primary objective of this project is to characterize the lifetime of three key components of the vacuum cleaner: the battery, the IR sensor, and the impact sensor. Component lifetimes may depend on various factors, including environmental properties (e.g., pet ownership, carpet score) and usage intensity (e.g., total cleaning time).

Careful consideration must be given to the modeling assumptions, particularly the censoring and truncation mechanisms inherent in the data collection process. All choices and assumptions will be clearly justified.

## 1. Baseline Suruvival Model: Kaplan Meier

We implement a Kaplan-Meier estimator for each component  battery, impact sensor, and IR sensor. The Kaplan-Meier estimator provides a non-parametric estimate of the survival function, allowing us to determine the probability that a component survives beyond a given time while accounting for right-censored observations.

```python
cols = ['Battery status', 'Impact status', 'IR status']
df[cols] = df[cols].fillna(0)
df.head()
```

```python
components = {
    "Battery status": "Battery",
    "IR status": "IR Sensor",
    "Impact status": "Impact Sensor"
}

fig, axes = plt.subplots(2, 2, figsize=(12,10), sharey=True)
axes = axes.flatten()
for ax, (col, label) in zip(axes, components.items()):
    kmf = KaplanMeierFitter()
    kmf.fit(durations=df["Time"], event_observed=df[col], label=label)
    kmf.plot_survival_function(ax=ax)
    if label == 'Battery':
        print("prediction ", kmf.predict(500), " for ", label)
    print("prediction ", kmf.predict(1750), " for ", label)
    ax.set_title(f"{label} Survival Curve")
    ax.set_xlabel("Time to Event (days)")
    ax.set_ylabel("Survival Probability")
fig.delaxes(axes[-1])

plt.tight_layout()
plt.show()
```

## 3) Inference section

## Sent for repair VS Components VS Total usage time

```python
def safe_km_quantile(kmf, q):
    try:
        val = kmf.percentile(q)
        return round(val, 2)
    except Exception:
        return "> observation period"

kmf = KaplanMeierFitter()

components = {
    "Battery status": "Battery Survival vs Total usage time",
    "Impact status": "Impact Survival vs Total usage time",
    "IR status": "IR Survival vs Total usage time"
}

inference_summary = []

for ax, (col, title) in zip(axes, components.items()):
    # Subset relevant columns
    temp = df[["Sent for repair", col, "UsageGroup", "Time"]].copy()

    # Combine component and usage group into one label
    temp[f"{col}_UsageGroup"] = temp[col].astype(int).astype(str) + "_" + temp["UsageGroup"].astype(str)

    for group in temp[f"{col}_UsageGroup"].unique():
        mask = temp[f"{col}_UsageGroup"] == group
        kmf.fit(
            temp["Time"][mask],
            temp["Sent for repair"][mask],
            label=group
        )

        median_time = kmf.median_survival_time_
        l10_time = safe_km_quantile(kmf, 0.10)

        median_display = round(median_time, 2) if np.isfinite(median_time) else "Not reached"
        l10_display = round(l10_time, 2) if isinstance(l10_time, (int, float)) and np.isfinite(l10_time) else "Not reached"

        total_units = mask.sum()
        failures = int(temp.loc[mask, "Sent for repair"].sum())
        fail_rate = f"{(failures / total_units) * 100:.2f}%" if total_units > 0 else "NA"

        usage_group = group.split("_")[-1]
        component_name = title.replace(" Survival vs Total usage time", "").split()[0]

        # append to dictionnary
        inference_summary.append({
            "Component": component_name,
            "UsageGroup": usage_group,
            "Total Units": total_units,
            "Failures": failures,
            "Failure Rate": fail_rate,
            "Median Survival (days)": median_display,
            "L10 (days)": l10_display
        })


inference_usage_df = pd.DataFrame(inference_summary)
inference_usage_df = inference_usage_df.sort_values(["Component", "UsageGroup"]).reset_index(drop=True)

print("Inference Summary: Components vs Total Usage Time")
display(inference_usage_df)

print("\nNote: 'Not reached' indicates too few failures within the study period to estimate the metric.")
```

## Components VS Pets and Total usage time

```python
def safe_km_quantile(kmf, q):
    try:
        val = kmf.percentile(q)
        return round(val, 2)
    except Exception:
        return "> observation period"

kmf = KaplanMeierFitter()

components = {
    "Battery status": "Battery",
    "Impact status": "Impact Sensor",
    "IR status": "IR Sensor"
}

inference_summary = []

for col, comp_name in components.items():
    # Create a temporary dataframe for this component
    temp = df[["Time", col, "Pets", "UsageGroup"]].copy()

    for usage_level in temp["UsageGroup"].unique():
        for pets in temp["Pets"].unique():
            mask = (temp["UsageGroup"] == usage_level) & (temp["Pets"] == pets)

            if mask.sum() < 10:  # skip small subgroups
                continue

            kmf.fit(durations=temp.loc[mask, "Time"], event_observed=temp.loc[mask, col]
            )

            median_time = kmf.median_survival_time_
            l10_time = safe_km_quantile(kmf, 0.10)

            median_display = round(median_time, 2) if np.isfinite(median_time) else "Not reached"
            l10_display = round(l10_time, 2) if isinstance(l10_time, (int, float)) and np.isfinite(l10_time) else "Not reached"

            total_units = mask.sum()
            failures = int(temp.loc[mask, col].sum())
            fail_rate = f"{(failures / total_units) * 100:.2f}%" if total_units > 0 else "NA"

            # Append summary
            inference_summary.append({
                "Component": comp_name,
                "UsageGroup": usage_level,
                "Pets": pets,
                "Total Units": total_units,
                "Failures": failures,
                "Failure Rate": fail_rate,
                "Median Survival (days)": median_display,
                "L10 (days)": l10_display
            })

# Create final DataFrame
inference_usage_df = pd.DataFrame(inference_summary)
inference_usage_df = inference_usage_df.sort_values(["Component", "UsageGroup", "Pets"]).reset_index(drop=True)

print("Inference Summary: Component Lifetime by Usage and Environment")
display(inference_usage_df)
```

## Survival curves of the three components

```python
# change battery status, IR status, Impact status to Booleans

df["Battery status"] = (df["Battery status"]==1) & (df["Sent for repair"]==1)
df["IR status"]   = (df["IR status"]==1) & (df["Sent for repair"]==1)
df["Impact status"]  = (df["Impact status"]==1) & (df["Sent for repair"]==1)

df = df.dropna(subset=["Total usage time", "Carpet score"]).reset_index(drop=True)

print(f"{df[['Total usage time', 'Battery status', 'IR status', 'Impact status', 'Pets','Carpet score']].head()}")
```

```python
components = {
    "Battery": "Battery status",
    "IR": "IR status",
    "Impact": "Impact status"
}

OUT = "dist_plots"; os.makedirs(OUT, exist_ok=True)
kmf = KaplanMeierFitter()
EPS = 1e-3

for comp, evt in components.items():
    dsub = df[["Total usage time", evt]].dropna().copy()
    n0 = int((dsub["Total usage time"] <= 0).sum())
    if n0 > 0:
        print(f"[{comp}] adjusted {n0} non-positive usage values to {EPS}")
        dsub.loc[dsub["Total usage time"] <= 0, "Total usage time"] = EPS

    print(f"\n{comp} and  n={len(dsub)}")

    wb = WeibullAFTFitter().fit(dsub, "Total usage time", event_col=evt)
    ln = LogNormalAFTFitter().fit(dsub, "Total usage time", event_col=evt)
    ll = LogLogisticAFTFitter().fit(dsub, "Total usage time", event_col=evt)

    aics = {"Weibull": wb.AIC_, "Log-normal": ln.AIC_, "Log-logistic": ll.AIC_}
    best = min(aics, key=aics.get)
    print("AICs:", aics, "| Best fit:", best)

    if comp == "Battery":
        model = wb; label = "Weibull"
    elif comp == "IR":
        model = ln; label = "Log-normal"
    elif comp == "Impact":
        model = ll; label = "Log-logistic"

    fig, ax = plt.subplots()
    kmf.fit(dsub["Total usage time"], dsub[evt], label="Empirical KM").plot(ax=ax, ci_show=False)
    model.predict_survival_function(dsub).mean(axis=1).plot(ax=ax, label=label)
    ax.set_title(f"{comp}: KaplanMeier vs {label}")
    ax.set_xlabel("Total usage time (hours)")
    ax.set_ylabel("Survival probability")
    ax.legend()
    fig.tight_layout()
    fig.savefig(os.path.join(OUT, f"{comp}_KM_vs_{label.split()[0]}.png"), dpi=160)
    plt.show()
```
